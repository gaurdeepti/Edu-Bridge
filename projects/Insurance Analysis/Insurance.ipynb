#Visualization

```python
# darkgrid style
sns.set_style('darkgrid')
```

#visualization on continuous variable

```python
sns.kdeplot(df['age'],color='blue',shade=True)
plt.show()
```

```python
sns.distplot(df['bmi'])
plt.show()
```

```python
sns.distplot(df['charges'],kde=False,color='green',bins=15)
plt.show()
```

#visualization on Categorical plot

```python
sns.countplot(x='sex',data=df)
plt.show()
```

```python
sns.countplot(x='children',data=df)
plt.show()
```

```python
sns.countplot(x='smoker',data=df)
plt.show()
```

```python
sns.countplot(x='region',data=df)
plt.show()
```

#to check outliers in dataset

```python
sns.boxplot(df['age'])
plt.show()
```

```python
sns.boxplot(y=df['bmi'],color='yellow')
plt.show()
```

```python
sns.boxplot(df['charges'],color='blue')
plt.show()
```

```python
#checking relationship between the columns without hue)
sns.pairplot(df,vars=['age','bmi','charges','children'])
plt.show()
```

```python
#Sex as hue
sns.pairplot(df,vars=['age','bmi','charges','children'],hue='sex')
plt.show()
```

```python
#Smoker as hue
sns.pairplot(df,vars=['age','bmi','charges','children'],hue='smoker')
plt.show()
```

```python
#Region as hue
sns.pairplot(df,vars=['age','bmi','charges','children'],hue='region')
plt.show()
```

#comparing two variables

```python
sns.relplot(x='charges', y='bmi',col='region',hue='sex',kind='line',data=df)
plt.show()
```
