# How much marks females got in Math Subject?

```python
stu1=stu[stu['gender']=="female"]
stu1.sort_values('math score',ascending=False).head()
```

# How much marks males got in Math subject?

```python
stu2=stu[stu['gender']=="male"]
stu2.sort_values('reading score',ascending=False).head()
```

# Observation For Math Subject
 In math subject Females got more 100 out of 100 than males. three female students got full score out of 100 and in male only one student got 100 out of 100. Based on this dataset we can say Female are more intelligent than males in math subject.

# What is the Total reading score in all of  race/ethnicity ?

```python
stu1 = stu.copy()
stu1.groupby(['race/ethnicity'])['reading score',].sum()
```

```python
stu1.groupby(['race/ethnicity'])['reading score',].sum().plot.bar(color='skyblue',figsize=(8,5))
plt.ylabel('Total Reading Score')
plt.show()
```

# What is Total sum of writing score in parental level of education and lunch?

```python
stu2=stu.copy()
stu2=stu2.pivot_table('writing score',columns=['parental level of education','lunch'],aggfunc='sum')
stu2
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='parental level of education',y='writing score',data=stu,hue='lunch',
            order=stu['parental level of education'].value_counts().index,palette='crest')
plt.show()
```

# What is the Maximum Score in Math according to extra test course?

```python
stu3 = stu.copy()
stu3.groupby(['test preparation course'])['math score',].max()
```

```python
stu3.groupby(['test preparation course'])['math score',].max().plot.bar(color='brown',figsize=(8,5))
plt.ylabel('Total math Score')
plt.show()
```
